```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities =pd.read_csv('./california_cities.csv')
```

```python
cities.head()
```

```python
lat=cities["latd"]
long=cities["longd"]
area, population=cities['area_total_km2'],cities['population_total']
```

```python
# 
plt.figure(figsize=(5,4))
plt.scatter(long, lat,
           c=np.log10(population),cmap='viridis',
           s=area, linewidths=0, alpha=0.5);
plt.axis('equal')
plt.xlabel('longtitude')
plt.ylabel('latitude')
plt.colorbar(label='log$_{10}$(population)')
plt.title('California Citie: Population and Area Distribution')
area_range=[50,100,300,500]
for area in area_range:
    plt.scatter([],[], s=area,
               label=str(area)+ 'km$^2$')
plt.legend()
```
